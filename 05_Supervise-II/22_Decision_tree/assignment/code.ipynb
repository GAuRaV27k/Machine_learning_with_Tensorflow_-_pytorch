```python
import pandas as pd 
from sklearn.model_selection import train_test_split 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("shop_smart_ecommerce.csv")
```

```python
df.head()
```

DA

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Count of Revenue by ProductRelated
sns.countplot(data=df, x='ProductRelated', hue='Revenue', ax=axes[0][0])
axes[0][0].set_title('Revenue by ProductRelated Pages')
axes[0][0].set_xlabel('ProductRelated')
axes[0][0].set_ylabel('Count')
axes[0][0].tick_params(axis='x', rotation=45)

# Revenue by VisitorType
sns.countplot(data=df, x='VisitorType', hue='Revenue', ax=axes[0][1])
axes[0][1].set_title('Revenue by Visitor Type')
axes[0][1].set_xlabel('VisitorType')
axes[0][1].set_ylabel('Count')

# Revenue by Month
sns.countplot(data=df, x='Month', hue='Revenue', ax=axes[1][0])
axes[1][0].set_title('Revenue by Month')
axes[1][0].set_xlabel('Month')
axes[1][0].set_ylabel('Count')
axes[1][0].tick_params(axis='x', rotation=45)

# Revenue by Weekend
sns.countplot(data=df, x='Weekend', hue='Revenue', ax=axes[1][1])
axes[1][1].set_title('Revenue by Weekend')
axes[1][1].set_xlabel('Weekend')
axes[1][1].set_ylabel('Count')

plt.tight_layout()
```
