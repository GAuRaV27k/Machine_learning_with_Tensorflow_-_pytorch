```python
import matplotlib.pyplot as plt 
from sklearn.model_selection  import train_test_split
from sklearn.svm import SVR
import pandas as pd 
from sklearn import datasets
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
df = datasets.load_diabetes(as_frame=True).frame
```

```python
df.head()
```

```python
X = df.drop("target",axis=1)
y = df["target"]
```

```python
# stratify=y ensures the class distribution in y is preserved in both train and test splits
X_train , X_test , y_train , y_test = train_test_split(
    X, y ,random_state=42,test_size=.30 
)
```

```python
y_scaler = StandardScaler()

y_train_scaled = y_scaler.fit_transform(y_train.values.reshape(-1,1)).ravel()
y_test_scaled = y_scaler.fit_transform(y_test.values.reshape(-1,1)).ravel()
```

```python
model = SVR()

model.fit(X_train,y_train_scaled)
```

```python
y_pred_scaled = model.predict(X_test)

print("r2Score :",r2_score(y_test_scaled, y_pred_scaled))
# r2_score(y_test,y_pred)
```
