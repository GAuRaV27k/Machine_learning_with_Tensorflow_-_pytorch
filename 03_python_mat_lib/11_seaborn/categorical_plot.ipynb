```python
import seaborn as sns

import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.barplot(
    data= tips,
    x= 'day',
    y= 'tip',
    hue='sex'
    
)
```

```python
sns.boxplot(
    data= tips,
    x= 'day',
    y= 'tip',
    hue='sex'
    
)
```
