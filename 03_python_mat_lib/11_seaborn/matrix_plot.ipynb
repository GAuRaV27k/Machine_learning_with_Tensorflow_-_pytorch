```python
# Seaborn

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
flight = sns.load_dataset('flights')
```

```python
flight.head()
```

HEATMAP

```python
flight_pivot = flight.pivot(index= 'month',columns='year',values='passengers')
```

To format annotations in a Seaborn heatmap, use the fmt parameter within the sns.heatmap() function, in conjunction with setting annot=True. The fmt parameter accepts standard Python string formatting codes. 

fmt='d': Integer format (no decimal places). Useful for counts or discrete data.
fmt='.1f': One decimal place for floating-point numbers.
fmt='.2f': Two decimal places for floating-point numbers.

```python
sns.heatmap(
    data=flight_pivot,
    cmap='mako',

    annot=True, # value of each box 
    fmt='d'# 
)
# plt.figure(figsize=(30, ))
plt.tight_layout()
```
