# Box Plots

A box plot is a statistical visualization that summarizes the distribution of a dataset using five key numbers:
- **Minimum** (lowest non-outlier)
- **First Quartile (Q1)** - 25th percentile
- **Median (Q2)** - 50th percentile
- **Third Quartile (Q3)** - 75th percentile
- **Maximum** (highest non-outlier)

Box plots are useful for identifying outliers, comparing distributions across groups, and understanding data spread and skewness.

```python
np.random.seed(0)
data = np.clip(np.random.randint(0, 100, 100),15,100)
data = list(data)
data.append(200)  # Adding an outlier
print(data)
plt.boxplot(data)
plt.grid(True,color='black', linestyle='--', linewidth=0.5,alpha=0.4)
```
