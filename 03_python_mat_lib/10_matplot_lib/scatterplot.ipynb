```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import string

age = np.random.randint(20, 80, 50)
blood_pressure = np.random.randint(110, 160, 50)
names = [
    string.ascii_uppercase[i % 26] + (str(i // 26 + 1) if i >= 26 else '')
    for i in range(len(age))
]
```

```python
plt.scatter(age,blood_pressure, color = 'red')
```

```python
colors = ['red', 'blue', 'green', 'yellow', 'purple', 'orange', 'pink', 'brown', 'gray', 'cyan' ]
plt.scatter(age,blood_pressure, color = colors)
```

logical coloring 

```python
colors = ['red' if bp > 135 else 'green' for bp in blood_pressure]
plt.scatter(age,blood_pressure, color = colors)
```

sizes 

```python
colors = ['red', 'blue', 'green', 'yellow', 'purple', 'orange', 'pink', 'brown', 'gray', 'cyan' ]

plt.scatter(age,blood_pressure, color = colors,s=blood_pressure)  # the size of points is determined by blood pressure values
```

maker and transparency 
transoarency is denotes by the alpha value 


```python
colors = ['red' if bp > 135 else 'green' for bp in blood_pressure]
plt.scatter(age,blood_pressure, color = colors,s=blood_pressure,marker='>',alpha=0.5)  # size of points determined by blood pressure values
```

: use a colormap to map blood pressure values to colors, enabling a gradient that highlights higher pressures with deeper shades and lower pressures with lighter shades

```python
plt.scatter(age,blood_pressure,s=blood_pressure,cmap='OrRd',c=blood_pressure) 

#  cmap = type of map 
# c = color based on blood pressure values
```

```python
plt.scatter(age,blood_pressure,s=blood_pressure,cmap='OrRd',c=blood_pressure) 

#  cmap = type of map 
# c = color based on blood pressure values
plt.colorbar(label= "Blood pressure")  # to show the color scale
plt.xlabel("Age")
plt.ylabel("Blood Pressure")
plt.title("Age vs Blood Pressure Scatter Plot")
plt.grid()
```

annotate 
add the text to each point 

`plt.annotate(text , point_positon , text_position)`

```python
plt.scatter(age,blood_pressure,s=blood_pressure,cmap='OrRd',c=blood_pressure) 

#  cmap = type of map 
# c = color based on blood pressure values
plt.colorbar(label= "Blood pressure")  # to show the color scale
plt.xlabel("Age")
plt.ylabel("Blood Pressure")
plt.title("Age vs Blood Pressure Scatter Plot")
plt.grid()

for i, name in enumerate(names):
    plt.annotate(name, (age[i], blood_pressure[i]), textcoords="offset points", xytext=(0,10), ha='center')
```
