```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# Create sample data for subplot practice
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(-x/10)
```

`The np.linspace function in the NumPy library for Python is used to create an array of evenly spaced numbers over a specified interval`<br>


`numpy.linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None)`

```python
plt.subplot(2, 2, 1)
plt.plot(x,y1)
plt.title('Sine Function')
plt.grid(True,color='black', linestyle='--', linewidth=0.5,alpha=0.4)
plt.subplot(2, 2, 2)
plt.plot(x,y2, 'r')
plt.title('Cosine Function')
plt.grid(True,color='black', linestyle='--', linewidth=0.5,alpha=0.4)
plt.subplot(2, 2, 3)
plt.plot(x,y3, 'g')
plt.title('Tangent Function')
plt.ylim(-10,10)
plt.grid(True,color='black', linestyle='--', linewidth=0.5,alpha=0.4)
plt.subplot(2, 2, 4)
plt.plot(x,y4, 'm')
plt.title('Exponential Decay')
plt.grid(True,color='black', linestyle='--', linewidth=0.5,alpha=0.4)
plt.tight_layout()
plt.show()
```
