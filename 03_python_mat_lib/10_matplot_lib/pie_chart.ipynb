```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
expenses = ["Salaries", "Rent", "Marketing", "R&D", "Miscellaneous"]
amounts = [500, 150, 120, 100, 50]
plt.style.use(  'petroff10')
plt.pie(amounts, labels=expenses)
plt.title("Company Expenses Distribution")
plt.show()
plt.style.available
```

customization 
1) autopct = "%1.1f%%" - show the precentage of each label 
2) colors = for the costom color 
3) wedgeprops = define (edge color ,linewidth ,linestyle, alpha , fill)
4) explode - highlights a slice by pulling it out 
5) startangle = rotate the pie chart 
6) shadow = add 3d effect or makes the chart circular 

autopct

```python

plt.style.use(  'petroff10')
plt.pie(amounts, labels=expenses,autopct='%1.1f%%')
  # to show percentage on pie chart
plt.title("Company Expenses Distribution")
plt.show()
```

wedgepropas
wedgeprops = define (edge color ,linewidth ,linestyle, alpha , fill)

```python

plt.style.use(  'petroff10')
plt.pie(amounts, labels=expenses,autopct='%1.1f%%')
  # to show percentage on pie chart
wedgeprops = {
    'edgecolor': 'black',
    'linewidth': 2,
    'linestyle': '--',
    'fill':'False'
}
plt.pie(amounts, labels=expenses, autopct='%1.1f%%', wedgeprops=wedgeprops)
plt.title("Company Expenses Distribution")
plt.show()
```

explode 

```python

plt.style.use(  'petroff10')
plt.pie(amounts, labels=expenses,autopct='%1.1f%%')
  # to show percentage on pie chart
wedgeprops = {
    'edgecolor': 'black',
    'linewidth': 2,
    'linestyle': '--',
    
}
plt.pie(amounts, labels=expenses, autopct='%1.1f%%', wedgeprops=wedgeprops,explode=[0,0,0,0.2,0])
plt.title("Company Expenses Distribution")
plt.show()
```

startangle

```python

plt.style.use(  'petroff10')
# plt.pie(amounts, labels=expenses,autopct='%1.1f%%')
  # to show percentage on pie chart
wedgeprops = {
    'edgecolor': 'black',
    'linewidth': 2,
    'linestyle': '--',
    
}
plt.pie(amounts, labels=expenses, 
        autopct='%1.1f%%', wedgeprops=wedgeprops,explode=[0,0,0,0.2,0]
        ,startangle=90
        )
plt.title("Company Expenses Distribution")

```

shadow

```python
plt.style.use(  'petroff10')
# plt.pie(amounts, labels=expenses,autopct='%1.1f%%')
  # to show percentage on pie chart
wedgeprops = {
    'edgecolor': 'black',
    'linewidth': 2,
    'linestyle': '--',
    
}
plt.pie(amounts, labels=expenses, 
        autopct='%1.1f%%', wedgeprops=wedgeprops,explode=[0,0,0,0.2,0]
        ,startangle=90,
        shadow=True
        )
plt.title("Company Expenses Distribution")
```
