```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Temperature over cities
days = ["Mon", "Tue", "Wed", "Thu", "Fri"]

cities = ["New York", "London", "Delhi", "Tokyo"]

temperatures = [
    [22, 23, 21, 24, 25],   # New York
    [18, 19, 17, 20, 21],   # London
    [30, 32, 31, 33, 34],   # Delhi
    [25, 26, 24, 27, 28]    # Tokyo
]
```

```python
# plt.Subplot(2,2,1)
fig ,axes = plt.subplots(2,2)
city_no = 0
for i in range(2): # for row
    for j in range(2): # for columns
        axes[i][j].plot(days,temperatures[city_no],marker='o',mfc='red',mec='black')
        # plt.legend(loc= 'upper left')
        axes[i][j].set_title(cities[city_no])
        # plt.title('cites vs temperatures')
        # plt.xlabel("DAYS")
        # plt.ylabel("TEMPERATURE")
        city_no +=1

fig.supxlabel("DAYS", y=0.02)
fig.supylabel("TEMPERATURE", x=0.02)
fig.tight_layout()
```
