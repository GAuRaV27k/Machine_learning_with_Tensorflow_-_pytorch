```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
direct = [50, 60, 70, 80, 90, 100, 110]
organic = [30, 40, 50, 55, 60, 70, 80]
social = [20, 25, 30, 35, 40, 50, 60]

plt.stackplot(days, direct, organic, social, colors=['#ff9999','#66b3ff','#99ff99'],labels=['Direct','Organic','Social Media'])
plt.title('Website Traffic Sources Over a Week')
plt.xlabel('Days')
plt.ylabel('Number of Visitors')
# plt.show()
plt.legend(loc='upper left')
```
