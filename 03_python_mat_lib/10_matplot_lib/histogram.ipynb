```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# Generate 1000 random marks from a normal distribution with mean 70 and standard deviation 15
marks = np.random.normal(70, 15, 1000)
```

```python
plt.hist(marks, bins=10, edgecolor='black',color='orange')
plt.title('Histogram of Student Marks')
plt.xlabel('Marks')
plt.ylabel('Number of Students')
plt.grid(True,color='black', linestyle='--', linewidth=0.5,alpha=0.4)
```

```python
# Generate 1000 random marks from a normal distribution with mean 70 and standard deviation 15
np.random.seed(0)
marks = np.clip(np.random.normal(70, 15, 1000), 25, 100)  # Clip marks to range [25, 100]
```

```python
plt.hist(marks, bins=10, edgecolor='black',color='orange')
plt.title('Histogram of Student Marks')
plt.xlabel('Marks')
plt.ylabel('Number of Students')
bins = [25,40,55,70,85,100]
plt.grid(True,color='black', linestyle='--', linewidth=0.5,alpha=0.4)
```

plot multiple dataset 

```python
# Create two datasets: legitimate and fraudulent transaction amounts
np.random.seed(0)
legit = np.clip(np.random.normal(50, 20, 500), 5, 150)  # Legitimate transactions
fraud = np.clip(np.random.normal(100, 30, 200), 10, 200)  # Fraudulent transactions

# Plot both datasets
plt.figure(figsize=(10, 6))
plt.hist(legit, bins=bins, edgecolor='black', color='green', alpha=0.6, label='Legitimate')
plt.hist(fraud, bins=bins, edgecolor='black', color='red', alpha=0.6, label='Fraud')
plt.title('Histogram of Transaction Amounts: Legitimate vs Fraud')
plt.xlabel('Transaction Amount($)')
plt.ylabel('Number of Transactions')
plt.legend()
plt.grid(True, color='black', linestyle='--', linewidth=0.5, alpha=0.4)
plt.show()
```
