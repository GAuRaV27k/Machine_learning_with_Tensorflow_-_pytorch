```python
import numpy as np 
import pandas as pd 
from matplotlib import pyplot as plt
```

```python
df = np.random.normal(loc=70, scale=10, size=200)
```

```python
plt.hist(df,bins=10)
plt.title("Question 1 plot ")
plt.xlabel("Score")
plt.ylabel("Frequency")
```

```python
plt.hist(df, bins=10)
plt.title("Question 1 plot")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.axvline(x=df.mean(), color='red', linestyle='--', linewidth=2, label=f'Mean = {df.mean():.2f}')
plt.legend()
plt.show()

print(f"Mean: {df.mean():.2f}")
print(f"Standard Deviation: {df.std():.2f}")
print("\nInterpretation:")
print("The red dashed line shows the mean score is approximately 70, which matches our")
print("expected value (loc=70). The distribution appears roughly normal/bell-shaped,")
print("centered around 70 with most scores falling between 60-80 (within 1 standard deviation).")
```
