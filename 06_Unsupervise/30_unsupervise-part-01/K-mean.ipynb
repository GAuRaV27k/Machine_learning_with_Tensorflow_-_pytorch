```python
import seaborn as sns 
from sklearn.datasets import make_blobs  # create the data of cluster 
```

```python
X ,y = make_blobs(   # y is labels(cluster id )
    n_samples=50_000,
    n_features=2,
    centers=4,  # number of  true clusters 
    random_state=42,
)
```

```python
sns.scatterplot(x=X[:,0],y=X[:,1])    # [:] -> all , [:,1 or 0] which column of feature 
```

These algorithm is work on distance of we scale the data

```python
# K mean 

from sklearn.cluster import KMeans

k = 4 

Kmean = KMeans(
    n_clusters=4,
    init='k-means++', # initiatiese the centroid 
    random_state=42

)
```

```python
labels = Kmean.fit_predict(X)  # Used in unsupervised learning  
# labels is the cluster number 
```

```python
sns.scatterplot(x=X[:,0],y=X[:,1],hue=labels,palette="bright") 

# sns.scatterplot(x=X[:,0],y=X[:,1],c=labels)
```
