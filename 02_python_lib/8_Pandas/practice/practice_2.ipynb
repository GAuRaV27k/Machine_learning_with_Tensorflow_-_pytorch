```python
df = pd.read_csv("Titanic-Dataset.csv")
```

Display only columns:A. Name ,Sex ,Age ,Fare ,Survived

```python
df[["Name","Sex","Age","Fare","Survived"]]
```

Select passengers who:
- •Arefemale
- •Fare>30

```python
sorted_female = df[(df["Sex"]=="female" )& (df["Fare"]>30)]
sorted_female
```

#### Group by Pclass and compute:
- Survival rate
- Average Fare
- Average Age

```python
classificed = df.groupby("Pclass").agg(
    Survival_rate = ("Survived","mean"),
    Average_fare = ("Fare","mean"),
    Average_Age = ("Age","mean"),
)
classificed
```
