```python
import json
```

```python
#function to load the data
def load_data(filename):
    with open(filename, "r") as f:
        data = json.load(f)
    return data 
```

```python
data = load_data("store_data.json")
print(data)
print(type(data))
```

clear and structure the data

```python
def clear_data(data):
    text_to_num = {"one" :1,"two":2,"three":3,"four":4,"five":5}
    clean_data = []
    unique_user =set()
    for user in data:
        #clean rating 
        raw_rating = str(user['rating'].strip().lower())
        if (raw_rating in text_to_num):
            raw_rating =text_to_num[raw_rating]
        user['rating'] = raw_rating
        print(raw_rating)
        # print(user)
        user["rating"] = raw_rating    

        # missing value 
        raw_age = user.get("age")
        if(raw_age) == None:
            user["age"] =None

    # deduplication
        if user["name"].strip() in unique_user:
            continue
        unique_user.add(user["name"])
        clean_data.append(user)

    return clean_data
```

```python
clear_data(data)
```

get meaning full data

```python
def get_insight(data):
    total_rating = 0 
    for user in data:
        total_rating += float(user["rating"])

    print(f"Average rating is ={(total_rating)/len(data)}")   

    # percentage with poor rating 
    poor_rating = 0 
    for user in data:
        if float(user["rating"]) <= 3.00:
            poor_rating +=1

    print(f"User with poor rating = {poor_rating /len(data)*100}%")
 
```

```python
get_insight(data)       
```

## recommendation feature 


```python
def recommendation_system(data):
    recommendation =[]
    for user in data:
        cur_recom = {}
        cur_recom["name"]  = user["name"]
        if( float(user["rating"]) >=4.0):
            cur_recom["brand"] ="apple"
        else:
            cur_recom["brand"] = "samsung"
        recommendation.append(cur_recom)    


    return recommendation    
```

```python
recommendation_system(data)
```
