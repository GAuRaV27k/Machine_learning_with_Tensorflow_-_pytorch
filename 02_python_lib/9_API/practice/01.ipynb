```python
from bs4 import BeautifulSoup
import pandas as pd 
import requests
```

### from quotes , extract all quotes and author from all pages with the life tag 

```python
URL = 'http://quotes.toscrape.com/page/10/'
res = requests.get(URL)
```

```python
res.status_code
```

```python
with open("data1.html",'w',encoding='utf-8') as f :
    f.write(res.text)
```

```python
with open('data1.html','r',encoding='utf-8') as f :
    content = f.read()

soup = BeautifulSoup(content, 'html.parser')
```

```python
raw_quotes = soup.find_all('span', attrs={'itemprop': 'text'})
quote =[]

for q in raw_quotes:
    quote.append(q.get_text(strip=True))
```

```python
quote
```

```python
author = []

for q in raw_quotes:
    author.append(q.find_next('small').get_text(strip=True))
```

```python
author
```

```python
raw_tags = soup.find_all('meta',attrs={'itemprop':'keywords'})
tags = []
for t in raw_tags:
    tags.append(t['content'].split(','))

```

```python
tags
```

```python
quote_detail= []
for q ,a,t in zip(quote,author,tags):
    quote_detail.append([q,a,t])
```

```python
quote_detail
```

```python
df = pd.DataFrame(quote_detail,columns=["Quote","author","Tags"])
```

```python
df.to_csv(
    "Save.csv",
    mode="a",          # append
    index=False,
    header=False       # don't repeat column names
)
```
