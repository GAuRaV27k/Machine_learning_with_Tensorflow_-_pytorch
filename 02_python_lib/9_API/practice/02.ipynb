```python
book_detail = []

rating_map = {
    "One": 1,
    "Two": 2,
    "Three": 3,
    "Four": 4,
    "Five": 5
}

for page in range(1, 51):   # 1 → 50
    URL = f"https://books.toscrape.com/catalogue/page-{page}.html"
    print("Scraping:", URL)
    res = requests.get(URL)
    with open("data2.html",'w',encoding='utf-8') as f :
        f.write(res.text)

    with open("data2.html",'r',encoding='utf-8') as f :
        content = f.read()

    soup = BeautifulSoup(content,'html.parser')
    raw_title = soup.find_all('h3')
    title = []
    for t in raw_title:
        title.append(t.get_text(strip=True))
    
    price = []
    for p in raw_title:
        price.append(p.find_next('p').get_text(strip=True))


    rate = soup.find_all('article', class_="product_pod")
    ratings =[]

    for r in rate:
        out = r.find('p', class_="star-rating")["class"][1]
        ratings.append(rating_map[out])

    book_detail = []

    for t,p,r in zip(title,price,ratings):
        book_detail.append([t,p,r])

    df = pd.DataFrame(book_detail,columns=["Title","Price","Rating"])
    df.to_csv(
    'Save2.csv',
    mode="a",          # append
    index=False,
    header=False       # don't repeat column names
    )
```
