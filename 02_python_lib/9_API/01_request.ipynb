```python
from bs4 import BeautifulSoup 
```

```python

with open(r'E:\AIML_apna_college\9API\scrap.html', 'r', encoding='utf-8') as f:
    Html_content =f.read()
```

```python
soup = BeautifulSoup(Html_content, 'html.parser')
```

```python
all_h3 = soup.find_all('h3')
# all_h3
name = []
for h3 in all_h3:
    name.append(h3.get_text(strip=True))

name
```

```python
cost = []
for h3 in all_h3:
    cost.append(h3.find_next('p').getText(strip=True))
    print(cost)
```

```python
book_detail = []

for n, c in zip(name, cost):
    book_detail.append([n, c])

book_detail
```

```python
import pandas as pd 
df = pd.DataFrame(book_detail,columns=["Title","Cost"])
```
