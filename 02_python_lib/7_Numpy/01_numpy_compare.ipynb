## Vectorization & Broadcasting in NumPy


Vectorization and Broadcasting in NumPy are two of the most powerful features for fast numerical computations.
### Vectorization
<b>Vectorization means performing operations on entire arrays at once without explicit</b><br>
Python loops.<br>
- NumPy uses C-level implementations internally → much faster than Python loops.
- Makes code shorter, cleaner, and faster


### Broadcasting <br>
Broadcasting allows NumPy to automatically expand arrays of different shapes so<br>
that arithmetic operations can be performed. It’s basically scaling arrays without using<br>
extra memory.<br>
- No need to manually reshape arrays.
- Useful for combining arrays of different dimensions.
<br>
<b>Broadcasting Rules</b><br>
Broadcasting can only take place when the arrays are of compatible shape. So NumPy <br>
compares shapes of arrays from right to left. For the array to be compatible, all<br>
dimensions must either be:

- • Equal, or
- • 1, or
- • Missing (smaller array can be “stretched”

```python
arr = np.array([1, 2, 3, 4, 5])
sq_arr = arr ** 2 # Square of all nums
print(sq_arr)
arr2 = np.array([6, 7, 8, 9, 10])
print(arr + arr2) # Sum of 2 arrays
```

```python
arr_mul10 = arr * 10 # Multiply by 10 to all nums
print(arr_mul10)
# Broadcasting with a Vector
arr1D = np.array([1, 2, 3])
arr2D = np.array([[1, 2, 3], [4, 5, 6]])
print(arr1D + arr2D)
```

## Vector Normalization 

A quite common example of broadcasting in Vector Normalization. This is very<br>
common in machine learning and data preprocessing.<br>
Let’s take an example of Standard Vector Normalization i.e. transforming an array<br>
such that it has:<br>
- • mean = 0
- • standard deviation = 1
For each element xi in vector, xi<br>
normalized = (xi − μ)/σ<br>
Where:<br>
- • μ = mean of the vector (or column)
- • σ = standard deviatio

```python
# Standard Vector Normalization
arr = np.array([[1, 2], [3, 4]])
mean = np.mean(arr)
std_dev = np.std(arr)
normalized_arr = (arr - mean) / std_dev
print(normalized_arr)
# Column wise Normalization
arr = np.array([[1, 2], [3, 4], [5, 6]])
mean = np.mean(arr, axis = 0)
std_dev = np.std(arr, axis = 0)
print((arr - mean) / std_dev)
```
