```python
import torch
import torch.nn as nn 
```

```python
torch.manual_seed(42)  # random state of pytorch
neuron = nn.Linear(in_features=3,out_features=1)
```

```python
inputs = torch.tensor([1.0,2.0,3.0])
```

```python
output = neuron(inputs)
```

```python
print(output)
```

```python
neuron.weight
```

```python
neuron.bias
```

```python
# -- verify 

total = neuron.weight@inputs +neuron.bias

print(total)
```
