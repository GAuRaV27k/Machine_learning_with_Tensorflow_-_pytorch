```python
import pandas as pd 
import numpy as np 
```

```python
df = pd.read_csv("HousePricePrediction.csv")
```

```python
# Drop records with null values
# Drop Id as it doesn't contribute in price
df.drop(['Id'],
             axis=1,
             inplace=True)
df = df.dropna()

df.isnull().sum()
df
```

<!-- s physicalfeatures, location, and condition -->

```python
categories = df["MSZoning"].unique().tolist()
print(categories)
```

```python
X = df.drop(columns=['LotConfig','Exterior1st',"SalePrice","BsmtFinSF2"])
y = df["SalePrice"]


X['BldgType'] = X['BldgType'].map({'1Fam':1,'2fmCon':2,'Duplex':3,'TwnhsE':4,'Twnhs':5})
X['MSZoning']=X['MSZoning'].map({'RL':1, 'RM':2, 'C (all)':3, 'FV':4, 'RH':5, np.nan:6})

```

```python
X = X.fillna(X.mean())
y = y.fillna(y.mean())
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
y_test
```

```python
# evaluate the model

from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print(f"these is the evaluation value--{r2}" )
```

```python
n = X_test.shape[0]
p = X_test.shape[1]

adjusted_r2 = 1-((1-r2)*(n-1)/(n-p-1))

print(f"adjusted r2 {adjusted_r2}")
```
