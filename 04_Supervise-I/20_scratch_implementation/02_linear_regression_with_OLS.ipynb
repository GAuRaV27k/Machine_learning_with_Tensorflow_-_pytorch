```python
import numpy as np 
import pandas as pd 
```

Ordinary Least Square


```python
class LinearRegressionOLS:
    def __init__(self):
        self.bias = None 
        self.weights = None
    
    def fit(self, X, y):
        m = X.shape[0]
        X_update = np.c_[np.ones((m, 1)), X]  # ._c is using for concatenation 

        theta = np.dot(np.linalg.inv(np.dot(X_update.T, X_update)), np.dot(X_update.T, y))

        self.bias = theta[0]
        self.weights = theta[1:]

    def predict(self, X):
        y_pred = self.bias + np.dot(X, self.weights)
        return y_pred
```

```python
X = np.array([[1],[2],[3],[4],[5]])
y = np.array([2,4,6,8,10])


model = LinearRegressionOLS()
model.fit(X,y)

y_pred = model.predict(X)

print(y_pred)
```
