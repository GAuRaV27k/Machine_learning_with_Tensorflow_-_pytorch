```python
import pandas as pd 
import numpy as np 
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score,accuracy_score,recall_score
from sklearn.preprocessing import StandardScaler 
```

```python
df = pd.read_csv('Iris.csv')
```

```python
df.info()
```

```python
df.head()
```

### feature eng

```python
df['Sepal_LW_Ratio'] = round(df['SepalLengthCm']/df['SepalWidthCm'], 3)
df['Petal_LW_Ratio'] = round(df['PetalLengthCm']/df['PetalWidthCm'], 3)
df['Sepal_Petal_Ratio'] = round(df['SepalLengthCm']/df['PetalLengthCm'], 3)

```

```python
df
```

Logistic Regression


```python
X = df.drop(['Id','Species'],axis=1)
y = df['Species']
```

```python
X_train , X_test,y_train,y_test = train_test_split(
    X,y,test_size=0.20,random_state=42
)
```

```python
from sklearn.linear_model import LogisticRegression
logistic_model = LogisticRegression(max_iter=3000)
```

```python
logistic_model.fit(X_train,y_train)
```

precision_score in scikit-learn defaults to binary classification (one positive, one negative class). Since `y_test` and `y_pred` contain three classes (multiclass classification), precision must be aggregated across classes (e.g., with `average='weighted'`, `micro`, or `macro`). Without specifying an averaging strategy, sklearn cannot combine per-class precision values, so it raises an error.


```python
y_pred = logistic_model.predict(X_test)
```

```python
print("accuracy :-" , accuracy_score(y_test,y_pred) *100,"%")
print("precision :-" , precision_score(y_test,y_pred,average='weighted') *100,"%")
```

### navie_bayes

```python
from sklearn.naive_bayes import GaussianNB
```

```python
GaussianNB_model = GaussianNB()
GaussianNB_model.fit(X_train,y_train)
y_pred = GaussianNB_model.predict(X_test)
```

```python
print("accuracy :-" , accuracy_score(y_test,y_pred) *100,"%")
print("precision :-" , precision_score(y_test,y_pred,average='weighted') *100,"%")
```

### KNN classifier with pipe 

```python
from sklearn.pipeline import Pipeline
```

```python
from sklearn.neighbors import KNeighborsClassifier
kNN_model = KNeighborsClassifier()
```

```python
classifier = [('scaler', StandardScaler()), ('kNN', KNeighborsClassifier())]

pipe   = Pipeline(classifier)
```

```python
kNN_model.fit(X_train,y_train)
Y_pred = kNN_model.predict(X_test)
print("accuracy :-" , accuracy_score(y_test,Y_pred) *100,"%")
print("precision :-" , precision_score(y_test,Y_pred,average='weighted') *100,"%")
```

```python
lst = [x for x in range(1,100,2)]
print(lst)
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'n_neighbors' :[x for x in range(1,100,2)]}

classifierCV = GridSearchCV(
    kNN_model,
    param_grid,
    cv = 5 
)
```

```python
classifierCV.fit(X_train,y_train)
```

```python
y_pred = classifierCV.predict(X_test)
print('recall',recall_score(y_test,y_pred,average='weighted'))
print('accuracy_score' ,accuracy_score(y_test,y_pred) *100,"%")
print('precision_score',precision_score(y_test,y_pred,average='weighted') *100,"%")
res = pd.DataFrame(classifierCV.cv_results_)
print(classifierCV.best_params_)
```
